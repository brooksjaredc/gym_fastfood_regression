## I used google trend to compute the ratio of searches for 'gyms near me' to 'fast food near me' per US state. Then I collected a set of features for each state that I thought might be predictive of that search ratio. 

## I start by collecting and aggregating the data, then I try out a set of regression algorithms, then I use some advanced regression analysis to further explore and understand the regression and remove some less useful features, and finish with some cross-validation to properly evaluate the performance of the linear regression.

```python
import pandas as pd
import numpy as np
```

```python
avg_commute_time = pd.read_csv('avg_commute_time.csv', sep='\t', header=0) #2014-2018 https://www.indexmundi.com/facts/united-states/quick-facts/all-states/average-commute-time#map
avg_commute_time.columns = ['State', 'avg commute time']
```

```python
pct_bachelors = pd.read_csv('pct_bachelors.csv', sep='\t', header=0) 
pct_bachelors.columns = ['State', 'percent bachelor or higher']
```

```python
pct_without_health_insurance = pd.read_csv('pct_without_health_insurance.csv', sep='\t', header=0) 
pct_without_health_insurance.columns = ['State', 'percent without health insurance']
```

```python
median_household_income = pd.read_csv('median_household_income.csv', sep='\t', header=0) 
median_household_income.columns = ['State', 'median household income']
```

```python
median_gross_rent = pd.read_csv('median_gross_rent.csv', sep='\t', header=0) 
median_gross_rent.columns = ['State', 'median gross rent']
```

```python
population_per_sq_mile = pd.read_csv('population_per_sq_mile.csv', sep='\t', header=0) 
population_per_sq_mile.columns = ['State', 'population per square mile']
```

```python
pct_obese = pd.read_csv('pct_obese.csv', sep=':\s', header=None, engine='python') # https://www.livescience.com/50973-obesity-rates-full-rankings.html
pct_obese.columns = ['State', 'percent obese']
```

```python
pct_obese_and_pop = pd.read_csv('pct_obese_and_pop.csv', header=0) # https://worldpopulationreview.com/state-rankings/obesity-rate-by-state
# pct_obese_and_pop.columns = ['State', 'percent obese', 'population']
pct_obese_and_pop = pct_obese_and_pop.append({'State': 'District of Columbia', 'obesityRate': '24.7', 'Pop': 684498}, ignore_index=True)
```

```python
fast_food_per_capita = pd.read_csv('fast_food_per_capita.csv', sep=':\s', header=None, engine='python') # https://www.thrillist.com/news/nation/states-with-most-fast-food-restaurants-datafiniti
fast_food_per_capita.columns = ['State', 'fast food per capita']
fast_food_per_capita.loc[44, 'State'] = 'District of Columbia'
```

```python
state_gyms = pd.read_csv('state_gyms.csv', sep='\t', header=0) # https://www.exercise.com/learn/10-states-with-the-most-gyms/
people_per_gym = state_gyms[['State', 'People Per Gym']]
```

```python
annual_rainfall = pd.read_csv('rainfall_per_state.csv', sep='\t', header=0) # http://www.usa.com/rank/us--average-precipitation--state-rank.htm
```

```python
google_search_trends = pd.read_csv('google_search_trends.csv', header=None)
google_search_trends.columns = ['State', 'gyms near me', 'fast food near me']
google_search_trends['ratio'] = google_search_trends['gyms near me']/google_search_trends['fast food near me']
google_ratio = google_search_trends[['State', 'ratio']]
```

```python
state_stats = avg_commute_time.merge(pct_bachelors, on='State').merge(pct_without_health_insurance, on='State').merge(median_household_income, on='State').merge(median_gross_rent, on='State').merge(population_per_sq_mile, on='State').merge(pct_obese_and_pop, on='State').merge(fast_food_per_capita, on='State').merge(people_per_gym, on='State').merge(annual_rainfall, on='State').merge(google_ratio, on='State')
```

```python
state_stats.replace(',','', regex=True, inplace=True)
state_stats
```

## Now that we've collected the (hopefully) relevant information, let's try some regression

```python
X = state_stats[['avg commute time', 'percent bachelor or higher', 'percent without health insurance', 
                'median household income', 'median gross rent', 'population per square mile',
                'obesityRate', 'fast food per capita', 'People Per Gym', 'Average Precipitation inches']]
y = state_stats['ratio']

c = X.select_dtypes(object).columns
X[c] = X[c].apply(pd.to_numeric,errors='coerce').copy()
```

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 0)

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import LinearRegression

linreg = LinearRegression().fit(X_train_scaled, y_train)

print('R-squared score (training): {:.3f}'
     .format(linreg.score(X_train_scaled, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linreg.score(X_test_scaled, y_test)))
```

### Just simply feeding our scaled data to a basic linear regression model gives us decent results, which is promising

### Now to try some slightly more sophisticated methods

### At the end we will use cross-validation to properly evaluate our chosen model, since we only have 51 rows (50 states + DC)

```python
from sklearn.linear_model import Ridge

linridge = Ridge(alpha=1.0).fit(X_train_scaled, y_train)

print('R-squared score (training): {:.3f}'
     .format(linridge.score(X_train_scaled, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linridge.score(X_test_scaled, y_test)))
print('Number of non-zero features: {}'
     .format(np.sum(linridge.coef_ != 0)))
```

```python
print('Ridge regression: effect of alpha regularization parameter\n')
for this_alpha in [0, 1, 3, 10, 20, 50, 100, 1000]:
    linridge = Ridge(alpha = this_alpha).fit(X_train_scaled, y_train)
    r2_train = linridge.score(X_train_scaled, y_train)
    r2_test = linridge.score(X_test_scaled, y_test)

    print('Alpha = {:.2f} r-squared training: {:.2f}, r-squared test: {:.2f}\n'
         .format(this_alpha, r2_train, r2_test))
```

### Ridge regression is actually decreasing the performance of the linear model.

```python
from sklearn.preprocessing import PolynomialFeatures

print('Polynomial Features: effect of polynomial degree\n')
for this_degree in [1, 2, 3, 4, 5]:
    poly = PolynomialFeatures(degree=this_degree)
    X_train_scaled_poly = poly.fit_transform(X_train_scaled)
    X_test_scaled_poly = poly.transform(X_test_scaled)

    linreg = LinearRegression().fit(X_train_scaled_poly, y_train)
    poly_train = linreg.score(X_train_scaled_poly, y_train)
    poly_test = linreg.score(X_test_scaled_poly, y_test)
    
    print('Polynomial degree = {:.2f} \
r-squared training: {:.2f}, r-squared test: {:.2f}\n'
         .format(this_degree, poly_train, poly_test))
```

```python
print('Polynomial Features: effect of polynomial degree with Ridge Regression\n')
for this_degree in [1, 2, 3, 4, 5]:
    poly = PolynomialFeatures(degree=this_degree)
    X_train_scaled_poly = poly.fit_transform(X_train_scaled)
    X_test_scaled_poly = poly.transform(X_test_scaled)
    
    linridge = Ridge(alpha = 1.0).fit(X_train_scaled_poly, y_train)
    poly_train = linridge.score(X_train_scaled_poly, y_train)
    poly_test = linridge.score(X_test_scaled_poly, y_test)

    print('Polynomial degree = {:.2f} \
r-squared training: {:.2f}, r-squared test: {:.2f}\n'
         .format(this_degree, poly_train, poly_test))
```

```python
from sklearn.linear_model import Lasso

print('Polynomial Features: effect of polynomial degree with Lasso Regression\n')
for this_degree in [1, 2, 3, 4, 5]:
    poly = PolynomialFeatures(degree=this_degree)
    X_train_scaled_poly = poly.fit_transform(X_train_scaled)
    X_test_scaled_poly = poly.transform(X_test_scaled)

    linlasso = Lasso(alpha=0.003, max_iter = 10000).fit(X_train_scaled_poly, y_train)
    poly_train = linlasso.score(X_train_scaled_poly, y_train)
    poly_test = linlasso.score(X_test_scaled_poly, y_test)

    print('Polynomial degree = {:.2f} \
r-squared training: {:.2f}, r-squared test: {:.2f}\n'
         .format(this_degree, poly_train, poly_test))
```

```python
from sklearn.neural_network import MLPRegressor

mlpreg = MLPRegressor(hidden_layer_sizes = [10],
                             activation = 'relu',
                             alpha = 0.0001,   #0.0003,
                             solver = 'lbfgs',
                             max_iter=1000,
                             random_state=80085).fit(X_train_scaled, y_train)

print('R-squared train score: {:.3f}'
     .format(mlpreg.score(X_train_scaled, y_train)))
print('R-squared test score: {:.3f}'
     .format(mlpreg.score(X_test_scaled, y_test)))
```

```python
from sklearn.neighbors import KNeighborsRegressor

knnreg = KNeighborsRegressor(n_neighbors = 9, algorithm='auto').fit(X_train_scaled, y_train)


print('R-squared train score: {:.3f}'
     .format(knnreg.score(X_train_scaled, y_train)))
print('R-squared test score: {:.3f}'
     .format(knnreg.score(X_test_scaled, y_test)))
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 3)

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)




import xgboost as xgb
model=xgb.XGBRegressor(max_depth=3,eta=0.05,min_child_weight=4)

model.fit(X_train_scaled, y_train)

print('R-squared train score: {:.3f}'
     .format(model.score(X_train_scaled, y_train)))
print('R-squared test score: {:.3f}'
     .format(model.score(X_test_scaled, y_test)))
```

### Let's briefly discuss each feature and its effect on the target value

### 'median household income' has a positive coefficient, implying that higher income communities are more likely to have more people searching for gyms and less people searching for fast food

### 'median gross rent' has a negative coefficient, implying that places with lower rent are more likely to have more people searching for gyms and less people searching for fast food. My guess is that places with higher rent have a higher proportion of nice apartment complexes or buildings with included gyms so that those people will use those gyms, instead of searching for external gyms

### 'fast food per capita' has a negative coefficient, implying that places with more fast food places per capita are more likely to have less people searching for gyms and more people searching for fast food. This one seems obvious: in places that have higher demand for fast food, there are more fast food places. But there could also be the effect of more choices in fast food lead to more searches (i.e. if there are only 2 fast food places in your area, you would quickly become familiar with your choices, but if there are a dozen in your area, you might crave variety and are more likely to search for what is available)

### 'People Per Gym' has a positive coefficient, implying that places with more commercial gyms are more likely to have more people searching for gyms and less people searching for fast food. This can possibly be explained by the same reasons given for the last two features: if lot's of people live in apartment buildings with included gyms, there will be less demand for gyms, and if there are only one or two gyms in your area, you may already be familiar with your choices before need to search google.

### 'Average Precipitation inches' has a positive coefficient, implying that places where it rains more are more likely to have more people searching for gyms and less people searching for fast food. This one is interesting, and may be explained by average rainfall and gym-going being correlated with a  third (unincluded) variable. I know that coastal cities tend to have fitter and more active people, and they usually get more rainfall than inland areas, so this could be a possible explanation

### Lastly, given the small number of data points (only 51), I wanted to see how a Leave-One-Out cross-validation would perform. In order to compare it to the train/test split, I have to use MSE, since you can't compute an r^2 value for one data point at a time. So let's first see how a single train/test split performs by computing RMSE (RMSE is more interpretable than MSE since its in the same units as the target variable)

```python
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state = 4)

scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)

X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.metrics import mean_squared_error

linreg = LinearRegression().fit(X_train_scaled, y_train)

print('R-squared score (training): {:.3f}'
     .format(linreg.score(X_train_scaled, y_train)))
print('R-squared score (test): {:.3f}'
     .format(linreg.score(X_test_scaled, y_test)))

y_pred_train = linreg.predict(X_train_scaled)
y_pred_test = linreg.predict(X_test_scaled)

print('Root Mean Squared Error (training): {:.3f}'
     .format(np.sqrt(mean_squared_error(y_train, y_pred_train))))
print('Root Mean Squared Error (test): {:.3f}'
     .format(np.sqrt(mean_squared_error(y_test, y_pred_test))))
```

```python
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score

linreg = LinearRegression()

crossvalidation = KFold(n_splits=51, random_state=None, shuffle=False)

scores = cross_val_score(linreg, X_scaled, y, scoring="neg_mean_squared_error", cv=crossvalidation,
 n_jobs=1)

print("Folds: " + str(len(scores)) + ", RMSE: " + str(np.sqrt(np.mean(np.abs(scores)))) + ", STD: " + str(np.std(scores)))
```

### It seems like the RMSE is similar for the LOO CV than the train/test split, which gives me slightly more confidence in the results

### And with the RMSE of just over 0.1 given a range of target values of over 0.6, this linear regression did a reasonably good job, implying that the 5 chosen features are reasonably predictive of the target variable
